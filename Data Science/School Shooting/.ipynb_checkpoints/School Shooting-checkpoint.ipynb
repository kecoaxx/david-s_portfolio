```python
#imports
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#import the data, using "datass" as the variable
datass = pd.read_csv('C:\Desktop\Coding\Data sci\School Shooting\cnn-school-shooting-data-master\dataset.csv')

#to make sure the data is valid (no null)
datass.dropna().sort_index()
```

## Plot of race that does the shooting

```python
def racess(race):
#     datass[datass[race] > 5].index
    sns.lineplot(x = 'year', 
                y = race,
                hue = 'race',
                data = datass
               )
    my_yticks = range(1,20)
    plt.yticks(my_yticks)
    plt.show()
    return

def userinput ():
    while True:
        try:
            bruh = input('killed/injured/victims: ')
        except:
            print('bruh')
            continue
    
        if bruh == ['killed','injured','victims']:
            continue
        else:
            print('bruh')
        
        

inp = userinput()
racess(inp)
```
