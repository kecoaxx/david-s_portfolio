```python
#imports
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#import the data, using "datass" as the variable
datass = pd.read_csv('C:\Desktop\Coding\Data sci\School Shooting\cnn-school-shooting-data-master\dataset.csv')

#to make sure the data is valid (no null)
pd_datass = datass.dropna().sort_index()

pd_datass.head()
```

## Plot of race that does the shooting

```python
#this function is to plot race using seaborn, with variable 'race'
def racess(race):
    sns.catplot(
                x = race,
                hue = 'race',
                kind = 'count',
                data = pd_datass,
                legend = True
                )
    plt.show()
    return
```

```python
#this function is for input and filter fot input to be only 'killed','injured', and 'victims'
def userinput ():
    while True:
        try:
            bruh = input('killed/injured/victims: ')
            exc = ['killed','injured','victims']
            if bruh in exc:
                return bruh
            else:
                print('input must be either killed, injured, or victims')
        except:
            continue

#executing code
inp = userinput() #inserting input from userinput() to variable 'inp'
racess(inp) #using input in 'inp', variable 'race' is filled, hence plot is generated
```
