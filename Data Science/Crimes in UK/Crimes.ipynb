```python
#imports
import pandas as pd
import glob
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Get a list of all CSV files in a directory
path = 'C:\Desktop\Coding\Data Sci\Crimes in UK\Datasets\*.csv'
files = glob.glob(path)

# Create an empty list to hold the dataframes
dfs = []

# Loop through the files and read them into dataframes
for file in files:
    df = pd.read_csv(file)
    dfs.append(df)

# Concatenate the dataframes into a single dataframe
combined_df = pd.concat(dfs)

# Write the combined dataframe to a new CSV file
combined_df.to_csv('combined.csv', index=False)
```

```python
url = 'https://gitlab.com/sammysitorus23/datasets/-/raw/main/combined.csv'
df_crimes = pd.read_csv(url, on_bad_lines= 'skip')
df_crimes = df_crimes.drop('Context', axis=1)
df_crimes = df_crimes.dropna()
df_crimes.head()
```

```python
df_crimes['Reported by'].duplicated().count()
```

```python
# sort the 'Crime type' column by count from highest to lowest
crime_counts = df_crimes['Crime type'].value_counts().sort_values(ascending=False).loc[10:]

sns.catplot(
    x='Crime type',
    kind='count',
    data=df_crimes,
    order=crime_counts.index # use the sorted index as the order for the x-axis categories
)

plt.xticks(rotation=90)

# loop through the sorted crime types and their counts to add text labels to the bars
for index, value in enumerate(crime_counts):
    plt.text(index, value + 3000, 
             str(value), rotation=90,
            ha = 'center' )

plt.show()
```
